```python
# ============================================================
# ULTIMATE HEART DISEASE AI + XAI PIPELINE
# Covers EVERYTHING: EDA, ML, DL, AutoML, XAI, PDP, SHAP, LIME
# ============================================================

# INSTALL FIRST (run once)
# pip install pandas numpy matplotlib seaborn scikit-learn
# pip install xgboost lightgbm catboost shap lime autogluon
# pip install pytorch-tabnet tab-transformer-pytorch pytorch-tabular torch


# ============================================================
# IMPORTS
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,
    auc,
    confusion_matrix
)

# Classical ML
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    GradientBoostingClassifier,
    AdaBoostClassifier,
    VotingClassifier,
    StackingClassifier
)

from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Boosting
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Transformers
from pytorch_tabnet.tab_model import TabNetClassifier
from tab_transformer_pytorch import TabTransformer

# NODE
from pytorch_tabular import TabularModel
from pytorch_tabular.models import NodeConfig
from pytorch_tabular.config import DataConfig, TrainerConfig, OptimizerConfig

# AutoML
from autogluon.tabular import TabularPredictor

# Explainability
import shap
from lime.lime_tabular import LimeTabularExplainer

import warnings
warnings.filterwarnings("ignore")


# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv("heart_disease_uci.csv")

df.drop(["id","dataset"], axis=1, errors="ignore", inplace=True)

print("Dataset shape:", df.shape)


# ============================================================
# EDA
# ============================================================

print(df.info())
print(df.describe())

plt.figure(figsize=(10,6))
sns.countplot(x=df["num"])
plt.title("Target Distribution")
plt.show()

plt.figure(figsize=(12,10))
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(12,10))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap (Numeric Features Only)")

plt.show()
# AFTER label encoding and imputation

# ============================================================
# PREPROCESSING
# ============================================================

cat_cols = df.select_dtypes(include="object").columns.tolist()
num_cols = df.select_dtypes(include=["int64","float64"]).columns.tolist()
num_cols.remove("num")

# impute categorical
cat_imputer = SimpleImputer(strategy="most_frequent")
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# encode categorical
encoders = {}

for col in cat_cols:

    df[col] = df[col].astype(str).str.lower()

    le = LabelEncoder()

    df[col] = le.fit_transform(df[col])

    encoders[col] = le


# impute numeric
num_imputer = SimpleImputer(strategy="median")
df[num_cols] = num_imputer.fit_transform(df[num_cols])


# ============================================================
# FEATURE ENGINEERING
# ============================================================

df["risk_score"] = df["cp"] + df["ca"] + df["oldpeak"]

df["stress_index"] = df["oldpeak"] * df["exang"]

df["chol_age_ratio"] = df["chol"] / df["age"]


# ============================================================
# SPLIT
# ============================================================

X = df.drop("num", axis=1)
y = (df["num"] > 0).astype(int)

feature_names = X.columns

X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.2,stratify=y,random_state=42)


# ============================================================
# SCALE
# ============================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# ============================================================
# MACHINE LEARNING MODELS
# ============================================================

models = {

"Decision Tree": DecisionTreeClassifier(),

"Logistic Regression": LogisticRegression(max_iter=5000),

"Gaussian NB": GaussianNB(),

"KNN": KNeighborsClassifier(),

"SVM": SVC(probability=True),

"Random Forest": RandomForestClassifier(),

"Extra Trees": ExtraTreesClassifier(),

"Gradient Boosting": GradientBoostingClassifier(),

"AdaBoost": AdaBoostClassifier(),

"XGBoost": XGBClassifier(eval_metric="logloss"),

"LightGBM": LGBMClassifier(),

"CatBoost": CatBoostClassifier(verbose=0),

"MLP": MLPClassifier(max_iter=2000)

}


results = {}

print("\nTraining Classical Models\n")

for name, model in models.items():

    model.fit(X_train_scaled, y_train)

    pred = model.predict(X_test_scaled)

    acc = accuracy_score(y_test, pred)

    results[name] = acc

    print(name, acc)


# ============================================================
# TABNET
# ============================================================

tabnet = TabNetClassifier(verbose=0)

tabnet.fit(X_train_scaled, y_train)

results["TabNet"] = accuracy_score(
    y_test,
    tabnet.predict(X_test_scaled)
)


# ============================================================
# FT TRANSFORMER
# ============================================================

class FTTransformer(nn.Module):

    def __init__(self,input_dim):

        super().__init__()

        self.fc = nn.Linear(input_dim,128)

        encoder = nn.TransformerEncoderLayer(
            d_model=128,nhead=8,batch_first=True)

        self.transformer = nn.TransformerEncoder(encoder,3)

        self.out = nn.Linear(128,1)

    def forward(self,x):

        x=self.fc(x)

        x=x.unsqueeze(1)

        x=self.transformer(x)

        x=x.squeeze(1)

        return torch.sigmoid(self.out(x))


model_ft = FTTransformer(X_train_scaled.shape[1])

optimizer=optim.Adam(model_ft.parameters())

loss_fn=nn.BCELoss()

X_tensor=torch.FloatTensor(X_train_scaled)
y_tensor=torch.FloatTensor(y_train.values).unsqueeze(1)

for epoch in range(100):

    optimizer.zero_grad()

    out=model_ft(X_tensor)

    loss=loss_fn(out,y_tensor)

    loss.backward()

    optimizer.step()


pred=model_ft(torch.FloatTensor(X_test_scaled)).detach().numpy()

results["FTTransformer"]=accuracy_score(
    y_test,(pred>0.5).astype(int)
)

from pytorch_tabular import TabularModel
from pytorch_tabular.models import NodeConfig
from pytorch_tabular.config import (
    DataConfig,
    TrainerConfig,
    OptimizerConfig,
    ExperimentConfig
)

print("\nTraining NODE model...")

train_node = pd.DataFrame(X_train_scaled, columns=feature_names)
train_node["target"] = y_train.values

test_node = pd.DataFrame(X_test_scaled, columns=feature_names)
test_node["target"] = y_test.values


data_config = DataConfig(
    target=["target"],
    continuous_cols=feature_names.tolist()
)

model_config = NodeConfig(
    task="classification",
    num_layers=2,
    num_trees=128,
    depth=6
)

trainer_config = TrainerConfig(
    max_epochs=50,
    batch_size=64
)

optimizer_config = OptimizerConfig()

experiment_config = ExperimentConfig(
    project_name="node_heart",
    run_name="node_run",
    log_target="tensorboard"   # ✅ FIXED
)


node_model = TabularModel(
    data_config=data_config,
    model_config=model_config,
    optimizer_config=optimizer_config,
    trainer_config=trainer_config,
    experiment_config=experiment_config
)

node_model.fit(
    train=train_node,
    validation=test_node
)

pred_df = node_model.predict(test_node)

print("Prediction columns:", pred_df.columns)

pred = pred_df.iloc[:,0].values

# IMPORTANT FIX
pred = (pred > 0.5).astype(int)

from sklearn.metrics import accuracy_score

print("NODE Accuracy:", accuracy_score(y_test, pred))
# ============================================================
# AUTOML
# ============================================================

train_df=pd.DataFrame(X_train_scaled,columns=feature_names)
train_df["target"]=y_train.values

predictor=TabularPredictor(label="target").fit(train_df,time_limit=300)

print(predictor.leaderboard())


# ============================================================
# MODEL COMPARISON
# ============================================================

plt.figure(figsize=(12,6))

plt.bar(results.keys(),results.values())

plt.xticks(rotation=45)

plt.title("Model Accuracy Comparison")

plt.show()


# ============================================================
# SHAP EXPLAINABILITY
# ============================================================

best_model=models["CatBoost"]

explainer=shap.Explainer(best_model,X_train_scaled)

shap_values=explainer(X_test_scaled)

shap.plots.bar(shap_values)

shap.plots.beeswarm(shap_values)

shap.plots.waterfall(shap_values[0])


# ============================================================
# LIME
# ============================================================

lime=LimeTabularExplainer(
    X_train_scaled,
    feature_names=feature_names,
    class_names=["No Disease","Disease"],
    mode="classification"
)

exp=lime.explain_instance(
    X_test_scaled[0],
    best_model.predict_proba)

exp.show_in_notebook()


# ============================================================
# PDP MANUAL (WORKS ALWAYS)
# ============================================================

rf=RandomForestClassifier()

rf.fit(X_train_scaled,y_train)

for i,feature in enumerate(feature_names):

    values=np.linspace(
        X_train_scaled[:,i].min(),
        X_train_scaled[:,i].max(),50)

    pdp=[]

    for val in values:

        temp=X_train_scaled.copy()

        temp[:,i]=val

        pdp.append(
            rf.predict_proba(temp)[:,1].mean()
        )

    plt.figure()

    plt.plot(values,pdp)

    plt.title(f"PDP: {feature}")

    plt.show()


# ============================================================
# CORRELATION HEATMAP
# ============================================================

corr=pd.DataFrame(
    X_train_scaled,
    columns=feature_names).corr()

plt.figure(figsize=(12,10))

sns.heatmap(corr,annot=True,cmap="coolwarm")

plt.show()


# ============================================================
# PREDICTION FUNCTION
# ============================================================

def predict_patient(patient):

    dfp=pd.DataFrame([patient])

    for col in cat_cols:

        dfp[col]=encoders[col].transform(
            [str(dfp[col].iloc[0]).lower()])

    dfp[num_cols]=num_imputer.transform(dfp[num_cols])

    dfp["risk_score"]=dfp["cp"]+dfp["ca"]+dfp["oldpeak"]

    dfp["stress_index"]=dfp["oldpeak"]*dfp["exang"]

    dfp["chol_age_ratio"]=dfp["chol"]/dfp["age"]

    Xp=scaler.transform(dfp)

    prob=tabnet.predict_proba(Xp)[0]

    print("Prediction:",
          "Disease" if prob[1]>0.5 else "No Disease")

    print("Probability:",prob)


# ============================================================
# COMPLETE
# ============================================================

print("\nPIPELINE COMPLETE")
```
